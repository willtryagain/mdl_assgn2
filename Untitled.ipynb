```python
S = []
invalid = (2, 2)
for i in range(1, 4):
    for j in range(1, 5):
        S.append((i, j))
S.remove(tuple(invalid))
S.remove((3, 4))
S.remove((2, 4))

p = {}

actions = {
    'r': ['u', 'd'],
    'l': ['u', 'd'],
    'u': ['l', 'r'],
    'd': ['l', 'r']
}

def add_key(p, s_, s, a, val):
    key = (s_, s, a)
    if key not in p:
        p[key] = 0
    p[key] += val

def get_next_state(s, a):
    s_ = list(s)
    
    if a == 'r':
        s_[1] = s_[1] + 1
        if s_[1] > 4:
            s_ = s
    elif a == 'l':
        s_[1] = s_[1] - 1
        if s_[1] < 1:
            s_ = s
    elif a == 'u':
        s_[0] = s_[0] + 1
        if s_[0] > 3:
            s_ = s       
    elif a == 'd':
        s_[0] = s_[0] - 1
        if s_[0] < 1:
            s_ = s
    s_ = tuple(s_)
    if s_ == invalid:
        s_ = s
    return s_
pdict = {}    
    
for s in S:
    for a in actions.keys():
        s_ = get_next_state(s, a)
        assert s_ != invalid, print(a, s)
        add_key(p, s_, s, a, 0.8)
        if (s, a) not in pdict:
            pdict[(s, a)] = set()
        pdict[(s, a)].add(s_)

        for a_ in actions[a]:
            s_ = get_next_state(s, a_)
            assert s_ != invalid, print(a_, s)
            add_key(p, s_, s, a, 0.1)
            pdict[(s, a)].add(s_)
gamma = 1
R = -0.022
epsilon = 0.1
mdp = {
    'states': S,
    'actions': actions,
    'transition': p,
    'rewards': R,
    'discount': gamma,
    'error': epsilon
}
```
